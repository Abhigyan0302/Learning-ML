```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv("PropertyPrice.csv")
print(data.head())
```

```python
%matplotlib
plt.xlabel('area(sq.ft.)')
plt.ylabel('price(US$)')
plt.scatter(data.area,data.price,color='red',marker='*')


```

```python
reg=linear_model.LinearRegression()
reg.fit(data[['area']],data.price)
```

```python

print(reg.predict([[5000]]))
print(reg.coef_)
print(reg.intercept_)
```

```python
test=pd.read_csv('testPrice.csv')
print(test.head(3))
```

```python
print(reg.predict(test))
```

```python
testset=reg.predict(test)
test['prices']=testset

```

```python
test.to_csv("testPrice.csv",index=False)
```
